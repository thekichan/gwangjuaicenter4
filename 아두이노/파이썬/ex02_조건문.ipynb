# 조건문

#### if 문
- if 조건식 :
- (들여쓰기) 실행문장
- 주의사항
    - 조건식에는 논리형결과를 써줘야한다! (True, False)
    - 반드시 콜론!!!!
    - 반드시 들여쓰기!!! -> 조건식의 실행범위 영역을 표시
   

```python
# 조건식이 True 였을 경우
if True :
    print("실행문장실행")
```

```python
# 조건식이 False 였을 때는 아무실행도 하지않고 끝난다
if False :
    print("실행문장실행")
```

- 들여쓰기의 중요성!

```python
if True:
    print("조건식이 참이었을 때 실행문장")
print("if문 바깥쪽에 있는 문장")
```

```python
if False :
    print("조건식이 거짓이었을 때 실행문장")
print("if문 바깥쪽에 있는 문장")
# 들여쓰기가 되어있지 않으면 if 의 영향을 받지 않는다
```

```python
money = 9000
if money >= 10000:
    print("택시를 탄다")
```

#### if else 문
- if 조건식:
- (들여쓰기)실행문장1
- else :
- (들여쓰기)실행문장2
- 주의사항
    - if, else는 같은 선상에 있어야 한다!
    - 콜론, 들여쓰기!!
    - else는 참이 아니였을 모든경우를 포함하기때문에 조건식을 적지 않는다!!

```python
# 단순 if문과의 차이점: 조건식이 False였을 경우 실행문장의 유무차이!!
if False:
    print("실행문장1")
else:
    print("실행문장2")
```

```python
money = 9000
if money >= 10000:
    print("택시를 탄다")
else:
    print("버스를 탄다")
```

```python
num = int(input("정수 입력 >>"))
if num % 3 == 0 and num % 5 == 0:
    print("3과 5의 배수입니다.")
else:
    print("3과 5의 배수가 아닙니다.")
                
```

```python
#1. 상품의 개수 입력받기
count = int(input("사려는 상품의 개수를 입력하세요>> "))
#2 조건문을 사용해서 금액계산
# 11개 미만 -> 10000원
if count <11 :
    price = 10000 * count
# 11개 이상 -> 9000원
else:
    price = 9000 * count
print(f"가격은 {price}원 입니다.")
```

### if elif else 문
- 여러개의 조건을 사용할 때

- 주의
    - 반드시 else 문을 작성해야한다 XXXXX

```python
# 돈이 10000원 이상일 경우 택시를 탄다
# 돈이 10000원 미만 1000원 이상일 경우 버스를 탄다
# 돈이 1000원 미만일 경우 걸어간다
money = int(input())
if money >= 10000:
    print("택시를 탄다")
elif money >= 1000:
    print("버스를 탄다")
else:
    print("걸어간다")
```

```python
money = int(input("입력 금액 >>"))
menu = input("입력 메뉴 >>")
coke = 600
cidar = 800
water = 1000

if money >= 600 and menu == "콜라":
    change = money - coke
    print(f"잔액은 1000원 {change//1000}개, 500원{change%1000//500}개, 100원{change%1000%500//100}개 입니다")

elif money >= 800 and menu == "사이다":
    change = money - cidar
    print(f"잔액은 1000원 {change//1000}개, 500원{change%1000//500}개, 100원{change%1000%500//100}개 입니다")

elif money >= 1000 and menu == "물":
    change = money - water
    print(f"잔액은 1000원 {change//1000}개, 500원{change%1000//500}개, 100원{change%1000%500//100}개 입니다")
else: 
    print("잔액이 부족합니다")
```
