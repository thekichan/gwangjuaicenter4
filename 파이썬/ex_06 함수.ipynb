```python
def cal(num1, num2, op):
    """덧셈과 뺄셈을 계산하는 함수"""
    if op == "+" :
        result = num1 + num2
    elif op == "-" :
        result = num1 - num2
    return result     

num1 = int(input("첫번째 정수 입력 >> "))
num2 = int(input("두번째 정수 입력 >> "))
op = input("연산자 입력(+,-) >> ")
result = cal(num1, num2, op)
print("결과 : ", result)
    
    
```

```python
li= [2, 7, 3, 5, 1, 2]
def isDuplicate(x):
    for l in range(0, len(li)-1):
        for i in range(l+1, len(x)):
            if x[l] == x[i]:
                return True
    return False

def createRandomNumber(cnt):
    randomList = [0,0,0]
    while isDuplicate(randomList):
        randomList = []
        for i in range(cnt):
            num = random.randint(1,5)
            randomList.append(num)
    return randomList
    
```

```python
def strikeBallCounting(li1, li2):
    strike = 0
    ball = 0
    
    for k in range(len(li1)):    
        for i in range(len(li2)) :
            if li1[k] == li2[i] :
                if k == i :
                    strike += 1
                else :
                    ball += 1
    return strike, ball

strikeBallCounting([1,2,3],[5,2,3])
```

```python

## print("===GameStart==")
randomList = createRandomNumber(3)
print(randomList)


while True :
    
    inputList = []
    for i in range(len(randomList)):
        num = int(input(f'{i+1}번째 숫자입력: '))
        inputList.append(num)
    print(inputList)

    strike, ball = strikeBallCounting(randomList, inputList)
    print(f"strike : {strike} \tball : {ball}")
    
    if strike == len(randomList) :
        print("게임종료!")
        break
```
