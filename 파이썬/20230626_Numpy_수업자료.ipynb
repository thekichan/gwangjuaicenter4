```python
rand = np.random.randint(1, 10, size = (2, 5))
rand
```

```python
# 배열에 있는 모든 값 더하기
print(np.sum(rand))
```

```python
# axis = 0: 축 변경 > 1차원 데이터를 합침
# axis = 1: 축 변경 > 2차원 데이터를 합침
print(np.sum(rand, axis = 0))
print(np.sum(rand, axis = 1))
```

```python
print(np.mean(rand, axis = 0))
print(np.mean(rand, axis = 1))
```

```python
# delimiter = 구분기호
data = np.genfromtxt('ratings.txt', delimiter = '::', dtype = 'int')
data
```

```python
data.shape
```

```python
# 1. 전체(1000209)명의 평점 평균 구하기
avg = np.mean(data[:, 2])
avg
```

```python
# 2. 1번 사람의 평점 평균 구하기
avg_1 = np.mean(data[data[:, 0] == 1, 2])
avg_1
```

```python
# 3. 모든 사람의 평점 평균 구하기
# 값을 리스트에 저장하기
lst = []

for i in range(1, 6041):
    lst.append([i, np.mean(data[data[:, 0] == i, 2])])
lst
```

```python
# 4. 각 사용자별 평균 평점이 4점 이상인 사용자 구하기
total_avg = np.array(lst)
bol = total_avg[:, 1] > 4
bol
total_avg[total_avg[:, 1] > 4, 0].astype(int)
```

```python
# 선생님 코드

# 1. 전체(1000209)명의 평점 평균 구하기
total_ratings = data[:, 2] # 전페 인원의 평점 데이터
print(np.mean(total_ratings), '\n')


# 2. 1번 사람의 평점 평균 구하기
# 불리언 인덱싱을 사용해서 1번 사용자의 데이터 가져오기
bol = data[:, 0] == 1
# 1번 사용자의 평점 데이터 가져오기
rating_1 = data[bol, 2]
# 평점의 평균 구하기
print(np.mean(rating_1), '\n')


# 3. 모든 사람의 평점 평균 구하기
# 값을 리스트에 저장
# 1번과 2번을 정리하기
total_list = []
for i in range(1, 6041):
    # bol = data[:, 0] == i
    # rating_1 = data[bol, 2]
    # mean_rating = np.mean(rating_1)
    # total_list.append([i, mean_rating])
# print(total_list, '\n')

# 한 줄 코드
    total_list.append([i, np.mean(data[data[:, 0] == i, 2])])
print(total_list, '\n')

# 4. 각 사용자별 평균 평점이 4점 이상인 사용자 구하기
total_np = np.array(total_list)
bol = total_np[:, 1] >= 4
print(total_np[bol, 0].astype(int)) # 타입 변경하기 astype(변경할 타입)
```
