```python
import pandas as pd
df2015 = pd.read_csv('./data/2015.csv',encoding = 'euc-kr', index_col = '관서명')
df2016 = pd.read_csv('./data/2016.csv',encoding = 'euc-kr', index_col = '관서명')
df2017 = pd.read_csv('./data/2017.csv',encoding = 'euc-kr', index_col = '관서명')

# 구분에 발생건수 데이터만 사용
# 총계 컬럼을 만들어야한다 > 구분 ~ 폭력까지의 값을 더하면 됨
# 2015 ~ 2017년 진행하기
# 2015 - 2016 증감률 구하기
# 2016 - 2017 증감률 구하기
# 전체 합치기
# df2015

# 데이터 총계 구하기
df2015['총계'] = df2015.loc[ : , '살인':'폭력'].sum(axis = 1)
df2016['총계'] = df2016.loc[ : , '살인':'폭력'].sum(axis = 1)
df2017['총계'] = df2017.loc[ : , '살인':'폭력'].sum(axis = 1)

# 쓸모없는 데이터 삭제하기
# unique() : 중복을 제거하고 한번만 출력
# df2015.index.unique()

# df2016.index.unique()

# df2017.index.unique()

# 광주지방경찰청 삭제
# drop(삭제할 인덱스)
df2017 = df2017.drop('광주지방경찰청')

# 원하는 데이터 가져오기
# 1. 구분 컬럼의 값이 발생건수인 데이터만 가져오기
bol = df2015['구분'] == '발생건수'
# df2015.loc[bol] 
# loc, iloc인덱서에 하나의 값을 입력하면 인덱스에 적용 모든 컬럼을 출력
s1 = df2015.loc[bol,'총계'] 
s1.name = '2015총계'

bol = df2016['구분'] == '발생건수'
s2 = df2016.loc[bol,'총계'] 
s2.name = '2016총계'

bol = df2017['구분'] == '발생건수'
s3 = df2017.loc[bol,'총계'] 
s3.name = '2017총계'

# 2015-2016 증감율
# (금년 - 작년) / 작년 * 100
s4 = (s2 - s1) / s1 * 100
s4.name = '2015-2016 증감율'

# 2016-2017 증감율 
s5 = (s3 - s2) / s2 * 100
s5.name = '2016-2017 증감율'

# 데이터 합치기
total = pd.concat([s1,s4,s2,s5,s3] , axis = 1)
total
```

```python
# 카테고리 데이터
# 0, 2, 10, 21, 23, 37, 31, 61, 20, ,41, 32, 100
# 값의 범위별로 새로운 값을 부여하고 싶을 때 사용하는 함수
```

```python
# 1 ~ 15 : 미성년자
# 16 ~ 25 : 청년
# 26 ~ 35 : 중년
# 36 ~ 60 : 장년
# 61 ~ 99 : 노년

ages =[0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 32, 100]
# 두 숫자 사이를 하나의 범위로 생각
# 첫번째 숫자 초과 두번째 숫자 이하의 범위를지정
bins = [0, 15, 25, 35, 60, 99]
# 범위에 부여하고 싶은 값 지정
labels = ['미성년자', '청년', '중년', '장년', '노년']
# pd.cut(데이터, 범위, 라벨) : 정해둔 범위의 값으로 데이터를 다시 나눔
result = pd.cut(ages, bins, labels = labels)
result
```

```python
# 데이터 프레임 합치는 함수 : merge
# value를 기준으로 데이터를 합침
# on = 'key' :  기준이 key 컬럼 안에 있는 value 들
# concat은 column과 index를 기준으로 합침
# concat과 merge의 차이 - 특이 concat에 join 같은 기준끼리 
pd.merge(df5,df6,on='key', how = 'inner')
```
