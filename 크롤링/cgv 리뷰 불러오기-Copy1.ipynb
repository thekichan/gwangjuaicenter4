```python
from selenium import webdriver as wb # ë¸Œë¼ìš°ì €(chrome, edge)
from selenium.webdriver.common.keys import Keys # í‚¤ë³´ë“œ
from selenium.webdriver.common.by import By # íŠ¹ì • ëŒ€ìƒ ì§€ì •
import time # íŠ¹ì • ë¶„ê¸°ì ë§ˆë‹¤ ì•½ê°„ì˜ ë”œë ˆì´ë¥¼ ì£¼ìž

# CGV ì˜í™” ê´€ëžŒí‰ ìˆ˜ì§‘
driver = wb.Chrome()
driver.get('https://www.cgv.co.kr/')

time.sleep(0.5)

btn_menu = driver.find_element(By.CSS_SELECTOR, '#cgvwrap > div.header > div.nav > div > ul > li:nth-child(1) > h2 > a')
btn_menu.click()

time.sleep(0.5)

btn_movie = driver.find_element(By.CSS_SELECTOR, '#contents > div.wrap-movie-chart > div.sect-movie-chart > ol:nth-child(2) > li:nth-child(2) > div.box-contents > a > strong')
btn_movie.click()

time.sleep(0.5)

for i in range(1,11):
    reviews = driver.find_elements(By.CSS_SELECTOR, '#movie_point_list_container>li>div.box-comment')

    for review in reviews:
        print(review.text)
    btn_number = driver.find_element(By.CSS_SELECTOR, f'#paging_point > li:nth-child({i}) > a')
    btn_number.click()
    time.sleep(0.5)

btn_next = driver.find_element(By.CSS_SELECTOR, '#paging_point > li:nth-child(11) > button')
btn_next.click()

while True:
    time.sleep(0.5)

    reviews = driver.find_elements(By.CSS_SELECTOR, '#movie_point_list_container>li>div.box-comment')

    for review in reviews:
        print(review.text)
        
    for i in range(3,13):
        time.sleep(0.5)

        reviews = driver.find_elements(By.CSS_SELECTOR, '#movie_point_list_container>li>div.box-comment')

        for review in reviews:
            print(review.text)
        
        btn_number = driver.find_element(By.CSS_SELECTOR, f'#paging_point > li:nth-child({i}) > a')
        btn_number.click()
        time.sleep(0.5)

    btn_next = driver.find_element(By.CSS_SELECTOR, '#paging_point > li:nth-child(13) > button')
    btn_next.click()
```
