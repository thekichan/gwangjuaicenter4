```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
```

```python
# 메뉴 이름
title = driver.find_elements(By.CSS_SELECTOR,'h4.h.fz_03')
for i in title:
    print(i.text)
```

```python
# 메뉴 가격
price = driver.find_elements(By.CSS_SELECTOR,'div.item-price > strong')
for i in price:
    print(i.text)
```

```python
# 더보기를 한번 클릭하기
# 더보기가 몇번 있는지 모르는상태
# 더보기가 없을때까지 클릭하기
while True :
    btn = driver.find_element(By.CSS_SELECTOR,'a.c_05')
    btn.click()
    time.sleep(0.5)
    
title = driver.find_elements(By.CSS_SELECTOR,'h4.h.fz_03')
for i in title:
    print(i.text)
    
price = driver.find_elements(By.CSS_SELECTOR,'div.item-price > strong')
for i in price:
    print(i.text)
```

```python
# 1. 한솥 메인페이지에서 시작
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/#none')
time.sleep(1)
# 2.1.1 닫기 클릭
# close = driver.find_element(By.CSS_SELECTOR,'span.btn-close > a')
# close.click()
# 2.1.2 화면 최대화하기
driver.maximize_window()

# menu에 마우스를 올린 상태에서 전체 메뉴 클릭해야 진행이 됨
# 지금은 menu클릭한번 전체메뉴 클릭 한번 하고있어서 넘어가지 않음
# 2.2 Menu 찾기
menu = driver.find_element(By.CSS_SELECTOR,'#gnb > div.gnb_menu > ul > li:nth-child(3) > p > a')

# 3. 전체 메뉴 찾기
total_menu = driver.find_element(By.CSS_SELECTOR,'#gnb > div.gnb_menu > ul > li:nth-child(3) > div > ul > li:nth-child(1) > a')

# 3.1 menu에 커서올리고 전체 메뉴 클릭하기
# 이벤트를 묶어서 실행하는 함수
from selenium.webdriver import ActionChains
# 실행해야할 동작을 나열
# 브라우저 연결 : ActionChains(driver)
# 마우스 올리기 : move_to_element(menu)
# 클릭 : click(total_menu)
# 실행하기 : perform()
ActionChains(driver).move_to_element(menu).click(total_menu).perform()
time.sleep(1)

# 4. 더보기 안나올때 까지 클릭
try:
    while True :
        btn = driver.find_element(By.CSS_SELECTOR,'a.c_05')
        btn.click()
        time.sleep(0.5)
except:
    print('더보기 클릭 완료')

# 5. 메뉴 이름 출력
title = driver.find_elements(By.CSS_SELECTOR,'h4.h.fz_03')
for i in title:
    print(i.text)

# 6. 메뉴 가격 출력
price = driver.find_elements(By.CSS_SELECTOR,'div.item-price > strong')
for i in price:
    print(i.text)
```
