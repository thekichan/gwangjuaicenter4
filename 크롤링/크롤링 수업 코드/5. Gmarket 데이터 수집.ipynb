```python
from selenium import webdriver as wb # 브라우저(chrome, edge)
from selenium.webdriver.common.keys import Keys # 키보드
from selenium.webdriver.common.by import By # 특정 대상 지정
import time # 특정 분기점(페이지가 변할때)마다 약간의 딜레이를 주자
```

```python
driver = wb.Chrome()
```

```python
driver.get("https://www.gmarket.co.kr/n/best")
```

```python
# By.CLASS_NAME : 클래스이름만 사용가능
# By.ID : 아이디만 사용가능
# By.CSS_SELECTOR : 태그, 클래스, 아이디 사용 가능
titles = driver.find_elements(By.CSS_SELECTOR, 'li a.itemname')
for title in titles:
    print(title.text)
```

```python
prices = driver.find_elements(By.CSS_SELECTOR,'li div.s-price strong span')
for price in prices:
    print(price.text)
```

```python
# 1. Gmarket Best에서 상위 10개 제품명과 가격 가져오기
driver = wb.Chrome()
driver.get("https://www.gmarket.co.kr/n/best")
prices = driver.find_elements(By.CSS_SELECTOR,'li div.s-price strong span')
titles = driver.find_elements(By.CSS_SELECTOR, 'li a.itemname')
for i in range(109):
    print('제품명 :', titles[i].text)
    print('가격 :',prices[i].text)
    print('*'*50)
```

```python
# 2. 제품을 직접 클릭해서 나온 페이지 에서 제품명과 가격 가져오기
# 클릭해서 페이지에 들어갔을때 > find_element(s) 변수들이 초기화된다
```

```python
# 페이지 변환시 변수 초기화
driver = wb.Chrome()
driver.get("https://www.gmarket.co.kr/n/best")
for i in range(10):
    # 몇번째 상품인지 선택하기위해 남겨둔것
    btns = driver.find_elements(By.CSS_SELECTOR, 'li a.itemname')
    btns[i].click()
    time.sleep(0.5)
    title = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic h1.itemtit')
    price = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic strong.price_real')
    print('제품명 :', title.text)
    print('가격 :',price.text)
    print('*'*50)
    driver.back() # btns 변수 값 초기화
    time.sleep(0.5)
driver.close()
```
