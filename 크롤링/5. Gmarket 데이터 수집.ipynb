```python
from selenium import webdriver as wb # 브라우저(chrome, edge)
from selenium.webdriver.common.keys import Keys # 키보드
from selenium.webdriver.common.by import By # 특정 대상 지정
import time # 특정 분기점(페이지가 변할 때)마다 약간의 딜레이를 주자
```

```python
driver = wb.Chrome()
driver.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9E%A5%EB%A7%88')
title = driver.find_elements(By.CLASS_NAME,'news_tit')
```

```python
for i in range(len(title)):
    print(title[i].text)
```

```python
driver = wb.Chrome()
driver.get('https://www.gmarket.co.kr/n/best')
title = driver.find_elements(By.CSS_SELECTOR, 'li a.itemname')
price = driver.find_elements(By.CSS_SELECTOR, 'li div.s-price strong span')
# 셀렉터를 사용하는 방법 - 여러가지 대상을 찾아오는 방법
for i in title:
    print(i.text)
```

```python
price = driver.find_elements(By.CSS_SELECTOR, 'li div.s-price strong span')
for k in price:
    print(k.text)
```

```python
for j in range(10):
    print('제품명 : ', title[j].text)
    print('가격 : ', price[j].text)
    print('*'*50)
```

```python
# 페이지 변환 시 변수 초기화
driver = wb.Chrome()
driver.get('https://www.gmarket.co.kr/n/best')
# 몇번째 상품인지 선택하기위해 남겨둔것

for i in range(10):
    btns = driver.find_elements(By.CSS_SELECTOR,'li a.itemname')
    btns[i].click()
    time.sleep(0.5)
    title2 = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic h1.itemtit')
    price2 = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic strong.price_real')
    print('제품명 :', title2.text)
    print('가격 :', price2.text)
    print('*'*50)
    driver.back()
    time.sleep(0.5)
driver.close()

```

```python
# 페이지 변환시 변수 초기화
driver = wb.Chrome()
driver.get("https://www.gmarket.co.kr/n/best")
for i in range(10):
    # 몇번째 상품인지 선택하기위해 남겨둔것
    btns = driver.find_elements(By.CSS_SELECTOR, 'li a.itemname')
    btns[i].click()
    time.sleep(0.5)
    title = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic h1.itemtit')
    price = driver.find_element(By.CSS_SELECTOR, 'div#itemcase_basic strong.price_real')
    print('제품명 :', title.text)
    print('가격 :',price.text)
    print('*'*50)
    driver.back() # btns 변수 값 초기화
    time.sleep(0.5)
driver.close()
```
