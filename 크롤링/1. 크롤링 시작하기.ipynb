```python
# 크롤링 : 데이터를 수집하는 행위
# 웹에서 원하는 데이터를 수집
# 웹은 html/css 언어로 디자인 되어있음
# html/css 에서 원하는 값을 수집
```

```python
# html은 tag로 이루어져 있다
# <div> </div> : 시작태그와 끝태그가 있다
# 시작태그와 끝태그 사이에 내용(content)를 담는다
# 크롤링에서는 주로 content(내용)을 수집
```

```python
# tag안에는 함수의 매개변수 같은 속성들이 있음
# class, id
# class : 동일한 대상을 여러번 표현할 때 사용
# 같은 이름이 여러개 존재
# id : 페이지에 한번 표현할 때 사용
# 같은 이름이 존재하지 않음
```

```python
# 원하는 태그에 접근하기 위해서
# 태그의 구조를 사용
# div1 아래 div2 아래 div3 아래 h2 아래 span을 가져와줘
# 태그의 class 또는 id를 사용해서 특정지어주기
```

```python
# 선택자
# 태그를 지칭할 수 있게 해줌
# div 아래
# 자식 선택자 : > : 하나만 건너갈 때
# 자손 선택자 : (띄어쓰기) : 여러개를 건너갈 때
# div > div > div > h2 > span
# div1 > h2 > span
```

```python
# div1.item h2#title > span
```

```python
# 서버에 값을 요청하기 위해서 사용
# 파이썬에서 브라우저(chrome, edge)역할
# 사이트에 요청을 하고 응답 정보를 처리
import requests as req


# 데이터를 수집할 주소
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9E%A5%EB%A7%88'

# 주소에 접근
# 해당 request(브라우저)에게 해당 url을 요청
# 클라이언트가 서버에게 요청(request)
# 서버가 클라이언트에게 응답(response) > html/css 문서를 가져옴
res = req.get(url)
# res : html/css 문서
```

```python
# 서버에 값을 요청하기 위해서 사용
# 파이썬에서 브라우저(chrome, edge)역할
# 사이트에 요청을 하고 응답 정보를 처리
import requests as req


# 데이터를 수집할 주소 / 방법 ctrl shift c
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9E%A5%EB%A7%88'

# 주소에 접근
# 해당 request(브라우저)에게 해당 url을 요청
# 클라이언트가 서버에게 요청(request)
# 서버가 클라이언트에게 응답(response) > html/css 문서를 가져옴
res = req.get(url)
# res : 응답받은 모든 것(객체)

res
# response 200 : 통신에 성공했습니다.
# response 400 : 클라이언트 오류(우리의 실수)
# 404 : 클라이언트가 요청한 문서를 찾지 못했다.
# = url 잘못 적었다.
# response 500 : 서버 오류(우리의 잘못은 없다)
```

```python
# BeautifulSoup : 글자정보를 html정보로 변경
from bs4 import BeautifulSoup as bs

#html/css 값 가져오기
soup = bs(res.text, 'lxml')
soup
```

```python
# 가져오고싶은 값 특정해서 가져오기
data = soup.select('a.news_tit')
# a태그의 class가 news_tit인것들을 가져와줘
# a.news_tit
data
```
