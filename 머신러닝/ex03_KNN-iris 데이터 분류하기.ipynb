# 3. 데이터 전처리
- 학습용 데이터 > 이미 전처리가 되어있음

## 4. 탐색적 데이터 분석
- 학습용 데이터 > 굳이 할 필요는 없음

## 5. 모델 선택 및 하이퍼 파라미터 튜닝

```python
# 데이터 분리
# 문제(X - iris['data'])와 정답(y - iris['target'])
X = iris['data']
y = iris['target']
# 훈련(train - 7)과 평가(test - 3) + 데이터 섞어주기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 6)
```

```python
# 데이터 분리 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# KNN모델 불러오기
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

## 6. 학습

```python
knn.fit(X_train,y_train)
```

# 7. 평가 및 예측

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

# 8. 하이퍼 파라미터 튜닝
- 5단계 : 모델 선택 및 하이퍼 파라미터 튜닝
- 6단계 : 학습
- 7단계 : 평가

```python
# 점수를 저장할 리스트 생성
train_list = []
test_list = []

# n_neighbors : 이웃의 수 결정 1 ~ 20

n_n_settings = range(1,21)

for n in n_n_settings:
    # 이웃수 조절하면서 모델 생성하기
    knn = KNeighborsClassifier(n_neighbors=n)
    
    # 모델 학습 : 6단계
    knn.fit(X_train, y_train)
    
    # 평가 : 7단계
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    
    # 평가 결과 저장
    train_list.append(train_score)
    test_list.append(test_score)
```

```python
train_list
```

```python
test_list
```

```python
# 그래프로 확인하기
import matplotlib.pyplot as plt
plt.plot(n_n_settings, train_list, label = 'Train', marker = 'o') # 꺽은선 그래프
plt.plot(n_n_settings, test_list, label = 'Test', marker = 'o')
plt.legend() # 각 그래프의 label표시하기
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.grid() # 격자무늬 표시
plt.show()
```
